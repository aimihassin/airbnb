### Question 4

The objective of this question is to find frequent itemsets using the dataset from Question 1. In the context of my dataset, I would like to know which house_type that airbnb guests frequently book when they are in Manhattan and/or Orlando. Knowing which house_type has highest booking frequency may be helpful for the host to understand about the behaviour of airbnb users. Besides, it is also may be beneficial for the airbnb users too as they able to know what kinds of house_type dominate a particular place.

#### Import Required Libraries

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

#### Load Data

```python
df = pd.read_csv('airbnb.csv')
df.sample(5)
```

```python
# Removing spaces from the 'house_type'
df['house_type'] = df['house_type'].str.strip()
df.sample(3)
```

#### Creating A List

```python
df1 = df[['house_type', 'city']]
df1
```

```python
house_list = df1.values.tolist()
print(house_list)
```

#### Converting List to DataFrame with Boolean Values

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(house_list).transform(house_list)
df2 = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df2
```

#### Find frequently occurring itemsets using Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_ap = apriori(df2, min_support=0.05, use_colnames=True)
```

```python
frequent_itemsets_ap
```

#### Find frequently occurring itemsets using F-P Growth

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets_fp=fpgrowth(df2, min_support=0.05, use_colnames=True)
```

```python
frequent_itemsets_fp
```

#### Mine the Association Rules

   1) **Support** = the popularity of item and can be calculated by finding the number of transactions containing a particular item divided by the total number of transactions.
    
   2) **Confidence** = the likelihood that an item B is also bought if item A is bought. It can be calculated by finding the number of transactions where A and B are bought together, divided by the total number of transactions where A is bought. Mathematically, it can be represented as: Confidence(A → B) = (Transactions containing both (A and B))/(Transactions containing A)
    
   3) **Lift** = Lift refers to the increase in the ratio of the sale of B when A is sold. Lift(A –> B) can be calculated by dividing Confidence(A -> B) divided by Support(B). Mathematically it can be represented as: Lift(A→B) = (Confidence (A→B))/(Support (B))

```python
from mlxtend.frequent_patterns import association_rules

rules_ap = association_rules(frequent_itemsets_ap, metric="confidence", min_threshold=0.7)
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.7)
```

```python
rules_ap
```

```python
rules_fp
```

**Comment:** From the results above, it is known that airbnb users tend to book 'entire apartment' or 'private room in apartment' when they are in Manhattan with confidence of 0.70 and 0.92 respectively. Meanwhile, in Orlando, the guests prefer to book 'entire house', 'entire condominium' or 'private room in house' with confidence ranging from 0.89 to 0.97. Here, we can understand that most of the booked airbnb listings in Manhattan are apartment-typed, while in Orlando the type of house would be a house or condominium.
